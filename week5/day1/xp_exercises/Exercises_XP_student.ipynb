### Réponses à l'Exercice 1 : Petits quiz

1.  **Quelle est la différence clé entre l’apprentissage automatique traditionnel et l’apprentissage profond ?**
    La différence clé réside dans la **représentation des données** et le **besoin d'ingénierie des fonctionnalités**. L'apprentissage automatique traditionnel nécessite une ingénierie manuelle des fonctionnalités par des experts pour extraire des caractéristiques pertinentes des données. L'apprentissage profond, en revanche, utilise des architectures de réseaux neuronaux multicouches (profondes) capables d'**apprendre automatiquement des représentations hiérarchiques des données** directement à partir des données brutes, réduisant ainsi le besoin d'ingénierie des fonctionnalités. Par exemple, pour la reconnaissance d'images, l'apprentissage traditionnel pourrait utiliser des descripteurs de formes et de textures définis manuellement, tandis que l'apprentissage profond apprendrait ces caractéristiques directement à partir des pixels de l'image.

2.  **Comment les réseaux neuronaux artificiels (RNA) imitent-ils le cerveau humain ?**
    Les RNA imitent le cerveau humain à un niveau très abstrait par leur **structure interconnectée et leur capacité d'apprentissage par l'expérience**. Comme les neurones biologiques, les neurones artificiels reçoivent des entrées, les traitent et produisent une sortie. Ces connexions ont des 'poids' qui sont ajustés (apprentissage) en fonction des données d'entraînement, un peu comme les synapses du cerveau se renforcent ou s'affaiblissent. Cependant, cette imitation est une simplification grossière et ne reproduit pas la complexité biologique ou fonctionnelle du cerveau.

3.  **Pourquoi l’apprentissage profond fonctionne-t-il mieux sur de grands ensembles de données comparé à l’apprentissage automatique traditionnel ?**
    L'apprentissage profond possède un **grand nombre de paramètres** et une **capacité de modélisation élevée**, ce qui lui permet de capturer des motifs complexes et des relations non linéaires dans de très grands ensembles de données. Les modèles traditionnels ont tendance à plafonner en performance après un certain volume de données car leur capacité à apprendre des représentations complexes est limitée. Les modèles profonds, avec plus de données, peuvent découvrir des fonctionnalités plus riches et plus abstraites, améliorant ainsi leur généralisation. Par exemple, un modèle de deep learning pour la détection d'objets peut exploiter des millions d'images pour apprendre à reconnaître des objets sous diverses conditions, ce qu'un algorithme SVM aurait du mal à faire sans une ingénierie de fonctionnalités massive.

4.  **Quels sont les défis du deep learning, et comment les relever ?**
    Les défis incluent la **nécessité de vastes quantités de données étiquetées**, la **complexité computationnelle** (exigeant du matériel puissant comme des GPU), la **difficulté d'interprétation** des modèles (boîte noire), et les problèmes de **surapprentissage**. On peut les relever par l'utilisation de **techniques d'augmentation de données** (pour les données), l'**optimisation des architectures** et l'utilisation de **ressources cloud** (pour la complexité), des **méthodes d'XAI (Explainable AI)** pour l'interprétation, et des techniques de **régularisation** (comme le dropout) pour le surapprentissage.

5.  **Qu’est-ce que l’ingénierie des fonctionnalités, et pourquoi n’est-elle pas nécessaire en apprentissage profond ?**
    L'ingénierie des fonctionnalités est le processus de **sélection, de transformation et de création de variables (caractéristiques) à partir de données brutes** pour améliorer les performances d'un modèle d'apprentissage automatique. Elle est cruciale pour l'apprentissage traditionnel. Elle n'est **pas nécessaire en apprentissage profond** car les réseaux neuronaux profonds, grâce à leurs multiples couches cachées, sont capables d'**apprendre et d'extraire automatiquement des représentations hiérarchiques des données**, de la plus simple à la plus complexe, sans intervention humaine directe.

6.  **Quel rôle jouent les couches cachées dans un modèle d’apprentissage profond ?**
    Les couches cachées sont les **principaux moteurs de l'apprentissage des représentations des données** dans un modèle d'apprentissage profond. Chaque couche cachée apprend à extraire des caractéristiques de plus en plus complexes et abstraites à partir de la sortie de la couche précédente. Les premières couches peuvent apprendre des motifs simples (comme les bords ou les textures dans les images), tandis que les couches plus profondes combinent ces motifs pour reconnaître des concepts plus élaborés (comme des parties d'objets ou des objets entiers).

7.  **Dans un réseau de neurones artificiel (RNA), quelle est la fonction d’une fonction d’activation ?**
    La fonction d'activation introduit la **non-linéarité** dans le réseau de neurones. Sans fonctions d'activation, un réseau neuronal, même avec de nombreuses couches, se comporterait comme un simple modèle linéaire, incapable d'apprendre des motifs complexes et non linéaires. Elles déterminent si un neurone doit être 'activé' (transmettre un signal) et l'intensité de cette activation, permettant ainsi au réseau de modéliser des relations complexes dans les données.

## 🌟 Exercise 3: Building a Simple Neural Network with TensorFlow or Keras

**As stated in the instructions**  
Build a classifier for MNIST with:
- One input layer
- One hidden layer with 128 neurons and ReLU activation
- One output layer with 10 neurons and softmax activation

Steps:
1. Import TensorFlow and needed modules from `tensorflow.keras`
2. Load the MNIST dataset
3. Normalize data by dividing by 255.0
4. One-hot encode the labels
5. Build a sequential model with Flatten, Dense(128, ReLU), Dense(10, softmax)
6. Compile with optimizer `adam`, loss `categorical_crossentropy`, metrics `accuracy`
7. Train and evaluate on the test set

**Guidance**  
You will execute prefilled boilerplate for imports and dataset loading. You will complete the one-hot encoding and the model specification, compile, train, and evaluate.

**Learning point**  
Normalization stabilizes optimization. One-hot encoding matches softmax outputs to target vectors. ReLU accelerates training by avoiding saturation, while softmax maps logits to a probability simplex.

![image.png](https://github.com/user-attachments/assets/61737b80-69cb-41a1-aea5-76092cd2a373)


```python
# PREFILLED: just execute
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.datasets import mnist

print("TensorFlow:", tf.__version__)

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Normalize to [0,1]
x_train = x_train.astype("float32") / 255.0
x_test  = x_test.astype("float32") / 255.0

# Flatten will be added in the model; keep images as 28x28 for now
print("Train shape:", x_train.shape, "Test shape:", x_test.shape)
```

```python
# To-Do: one-hot encode labels y_train and y_test using to_categorical
# Hints: use to_categorical
y_train_oh = to_categorical(y_train, num_classes=10)
y_test_oh  = to_categorical(y_test, num_classes=10)

print("Labels one-hot:", y_train_oh.shape, y_test_oh.shape)
```

```python
# To-Do:  Build a sequential model with Flatten, Dense(128, ReLU), Dense(10, softmax)
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax'),
])

model.summary()
```

```python
# To-Do: compile the model with adam, categorical_crossentropy, and accuracy
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# To-Do: train then evaluate
history = model.fit(x_train, y_train_oh, epochs=5, batch_size=32, validation_split=0.2)
test_loss, test_acc = model.evaluate(x_test, y_test_oh, verbose=0)
print({"test_loss": float(test_loss), "test_acc": float(test_acc)})
```

## 🌟 Exercise 5 (optional): Implementing Forward and Backward Propagation in Python

**As stated in the enoncé**  
You will code a simple neural network that performs forward propagation and backpropagation for a regression problem predicting exam scores from study hours.

**Guidance**  
Run the prefilled code. Then modify the learning rate or initial weights and observe changes in the loss and updates.
Provide a short written explanation.

**Learning point**  
Gradient descent updates parameters in the direction of steepest loss reduction.
The learning rate controls step size. Too small slows learning. Too large risks divergence.


```python
# PREFILLED: just execute
import numpy as np

# Initialize input data (features)
x = np.array([4, 80])  # 4 hours studied, previous test score: 80

# Initialize weights and bias
w = np.array([0.6, 0.3])  # Initial weights
b = 10  # Initial bias

# Forward Propagation
def forward_propagation(x, w, b):
    z = np.dot(x, w) + b  # Weighted sum
    return z  # Linear activation (No ReLU here, it's a regression task)

# Compute prediction
y_pred = forward_propagation(x, w, b)
y_true = 85  # Actual exam score

# Compute Loss (Mean Squared Error)
loss = 0.5 * (y_true - y_pred) ** 2

# Compute Gradients
grad_w = -(y_true - y_pred) * x  # Partial derivatives with respect to weights
grad_b = -(y_true - y_pred)  # Partial derivative with respect to bias

# Update Weights and Bias
learning_rate = 0.01
w_new = w - learning_rate * grad_w
b_new = b - learning_rate * grad_b

# Print Results
print("Initial Prediction:", y_pred)
print("Loss:", loss)
print("Updated Weights:", w_new)
print("Updated Bias:", b_new)
```

### Interprétation de l'Exercice 5 (optionnel)

L'algorithme de la **descente de gradient** réduit l'erreur en ajustant itérativement les poids et le biais du modèle dans la direction opposée au gradient de la fonction de perte. Le gradient indique la direction de la plus forte augmentation de l'erreur; par conséquent, se déplacer dans la direction opposée (négative du gradient) minimise l'erreur. Chaque mise à jour déplace les paramètres du modèle (poids et biais) vers un point où la fonction de perte est plus faible, cherchant à atteindre un minimum local ou global de la fonction de perte.

#### Effet de la modification du taux d'apprentissage ou des poids initiaux:

*   **Taux d'apprentissage (learning_rate) :**
    *   Si le `learning_rate` est trop **petit**, l'apprentissage sera très lent, et le modèle pourrait prendre beaucoup de temps pour converger vers la solution optimale, ou même se bloquer dans un minimum local. Le modèle ferait de très petits pas à chaque itération.
    *   Si le `learning_rate` est trop **grand**, le modèle pourrait **diverger**; c'est-à-dire que les poids et le biais pourraient fluctuer de manière erratique, voire augmenter, au lieu de converger. Le modèle pourrait "sauter" par-dessus le minimum de la fonction de perte.

*   **Poids initiaux (w) :**
    *   Des poids initiaux différents peuvent conduire à des **minima locaux différents** dans l'espace de la fonction de perte, car l'algorithme de descente de gradient peut converger vers le minimum le plus proche du point de départ. Cela peut influencer la performance finale du modèle. Pour cet exercice de régression simple, l'impact sera mineur, mais dans des réseaux plus complexes, cela peut être crucial.

En modifiant le `learning_rate` de `0.01` à `0.1` par exemple, on observerait une mise à jour des poids et du biais beaucoup plus importante à chaque pas. Si le `learning_rate` était réglé à `0.0001`, les ajustements seraient beaucoup plus subtils.
