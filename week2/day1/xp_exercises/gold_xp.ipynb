##  Exercise 1: Basic Data Analysis with Kaggle
**Dataset:** Boston Housing Dataset  
**Goal:** Compute mean, median, and standard deviation of key columns.

```python
# Load Boston Housing Dataset via sklearn (equivalent to Kaggle version)
from sklearn.datasets import fetch_california_housing

# Note: Boston dataset was deprecated; California Housing is the modern equivalent
housing = fetch_california_housing(as_frame=True)
df_housing = housing.frame

print('=== Housing Dataset: First Rows ===')
print(f'Shape: {df_housing.shape[0]} rows x {df_housing.shape[1]} columns')
print()
df_housing.head()
```

```python
# Basic statistical analysis on key columns
columns_to_analyze = ['MedHouseVal', 'MedInc', 'AveRooms', 'HouseAge']

print('=== Statistical Analysis: Mean, Median, Std Dev ===')
print()

stats = pd.DataFrame({
    'Column': columns_to_analyze,
    'Mean':   [round(df_housing[col].mean(), 4) for col in columns_to_analyze],
    'Median': [round(df_housing[col].median(), 4) for col in columns_to_analyze],
    'Std Dev':[round(df_housing[col].std(), 4) for col in columns_to_analyze],
    'Min':    [round(df_housing[col].min(), 4) for col in columns_to_analyze],
    'Max':    [round(df_housing[col].max(), 4) for col in columns_to_analyze],
})

stats
```

```python
# Visualization: Distribution of Median House Value
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram
axes[0].hist(df_housing['MedHouseVal'], bins=40, color='steelblue', edgecolor='white')
axes[0].axvline(df_housing['MedHouseVal'].mean(), color='red', linestyle='--', label=f"Mean: {df_housing['MedHouseVal'].mean():.2f}")
axes[0].axvline(df_housing['MedHouseVal'].median(), color='orange', linestyle='--', label=f"Median: {df_housing['MedHouseVal'].median():.2f}")
axes[0].set_title('Distribution of Median House Value')
axes[0].set_xlabel('Median House Value ($100k)')
axes[0].set_ylabel('Frequency')
axes[0].legend()

# Boxplot
axes[1].boxplot(df_housing['MedHouseVal'], vert=True, patch_artist=True,
                boxprops=dict(facecolor='steelblue', color='navy'))
axes[1].set_title('Boxplot: Median House Value')
axes[1].set_ylabel('Median House Value ($100k)')

plt.tight_layout()
plt.show()

print(f"\nKey Insight: The mean ({df_housing['MedHouseVal'].mean():.2f}) is higher than the median ({df_housing['MedHouseVal'].median():.2f}),")
print("indicating a right-skewed distribution with some high-value outliers.")
```

##  Exercise 2: Comparative Analysis — Retail Data
**Datasets:** Retail Dataset (structured) + Women's Clothing Reviews (unstructured)  
**Goal:** Compare insights from structured vs unstructured data.

```python
# --- Structured Data: Retail Dataset (simulated) ---

np.random.seed(42)
n = 200

df_retail = pd.DataFrame({
    'Transaction_ID': range(1001, 1001 + n),
    'Date': pd.date_range(start='2023-01-01', periods=n, freq='D'),
    'Store_ID': np.random.choice(['Store_A', 'Store_B', 'Store_C', 'Store_D'], n),
    'Category': np.random.choice(['Electronics', 'Clothing', 'Food', 'Home', 'Sports'], n),
    'Sales_Amount': np.round(np.random.uniform(10, 500, n), 2),
    'Quantity': np.random.randint(1, 20, n),
    'Customer_Age_Group': np.random.choice(['18-25', '26-35', '36-50', '50+'], n),
    'Payment_Method': np.random.choice(['Cash', 'Credit Card', 'Debit Card', 'Mobile Pay'], n)
})

print('=== Retail Dataset (Structured): First Rows ===')
print(f'Shape: {df_retail.shape}')
df_retail.head()
```

```python
# Analysis: Sales trends and store performance
print('=== Sales by Category ===')
print(df_retail.groupby('Category')['Sales_Amount'].agg(['mean', 'sum', 'count']).round(2))

print('\n=== Top Performing Stores ===')
print(df_retail.groupby('Store_ID')['Sales_Amount'].sum().sort_values(ascending=False))

print('\n=== Customer Age Group Distribution ===')
print(df_retail['Customer_Age_Group'].value_counts())
```

```python
# Visualizations for Retail Dataset
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Sales by Category
category_sales = df_retail.groupby('Category')['Sales_Amount'].sum().sort_values(ascending=False)
axes[0].bar(category_sales.index, category_sales.values,
            color=['#3498db','#e74c3c','#2ecc71','#f39c12','#9b59b6'])
axes[0].set_title('Total Sales by Category')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Total Sales ($)')
axes[0].tick_params(axis='x', rotation=15)

# Store performance
store_sales = df_retail.groupby('Store_ID')['Sales_Amount'].sum()
axes[1].pie(store_sales.values, labels=store_sales.index,
            autopct='%1.1f%%', startangle=140,
            colors=['#3498db','#e74c3c','#2ecc71','#f39c12'])
axes[1].set_title('Sales Distribution by Store')

plt.tight_layout()
plt.show()
```

```python
# --- Unstructured Data: Women's Clothing Reviews (simulated) ---

reviews_data = {
    'Review_ID': range(1, 21),
    'Rating': [5,4,2,5,3,4,1,5,4,3,5,2,4,5,3,4,5,1,4,5],
    'Review_Text': [
        'Absolutely love this dress! Perfect fit and amazing quality.',
        'Nice blouse, good material but a bit pricey for the quality.',
        'Very disappointed. The fabric feels cheap and the color faded after one wash.',
        'Best purchase ever! The dress is stunning and I received so many compliments.',
        'Decent quality but the sizing runs small. Order a size up.',
        'Pretty good overall. The stitching is neat and the color matches the photo.',
        'Terrible quality. Seams came apart after first wear. Not worth the money.',
        'Gorgeous top! Super comfortable and very flattering on all body types.',
        'Good but not great. Expected better material for the price point.',
        'Average quality. Nothing special but gets the job done for casual wear.',
        'Incredible dress! I wore it to a wedding and felt absolutely beautiful.',
        'Very poor quality. The buttons fell off immediately. Returning it.',
        'Lovely fabric and great fit. Will definitely order again from this brand.',
        'Perfect for summer! Light, airy and beautifully designed. Highly recommend.',
        'OK product. The color is slightly different from the website photo.',
        'Great value for money. Good quality stitching and comfortable to wear.',
        'Amazing skirt! The fabric is luxurious and it fits perfectly as described.',
        'Awful experience. Wrong size sent and the return process is a nightmare.',
        'Pretty satisfied. Minor issues with the zipper but overall good purchase.',
        'Wonderful addition to my wardrobe. Elegant and well-crafted piece!'
    ]
}

df_reviews = pd.DataFrame(reviews_data)
print('=== Clothing Reviews Dataset (Unstructured): First Rows ===')
df_reviews.head()
```

```python
# Basic sentiment analysis using keyword matching
positive_keywords = ['love', 'amazing', 'perfect', 'great', 'wonderful', 'incredible',
                     'gorgeous', 'beautiful', 'excellent', 'stunning', 'fantastic', 'best']
negative_keywords = ['terrible', 'awful', 'poor', 'disappointed', 'cheap', 'wrong',
                     'bad', 'worst', 'horrible', 'faded', 'fell off', 'nightmare']

def get_sentiment(text):
    """Classify sentiment based on keyword presence."""
    text_lower = text.lower()
    pos_score = sum(1 for word in positive_keywords if word in text_lower)
    neg_score = sum(1 for word in negative_keywords if word in text_lower)
    if pos_score > neg_score:
        return 'Positive'
    elif neg_score > pos_score:
        return 'Negative'
    else:
        return 'Neutral'

df_reviews['Sentiment'] = df_reviews['Review_Text'].apply(get_sentiment)

print('=== Sentiment Distribution ===')
print(df_reviews['Sentiment'].value_counts())
print()
print('=== Average Rating by Sentiment ===')
print(df_reviews.groupby('Sentiment')['Rating'].mean().round(2))
```

###  Comparative Analysis: Structured vs Unstructured

| Aspect | Structured (Retail Dataset) | Unstructured (Clothing Reviews) |
|---|---|---|
| **Format** | Tables with fixed rows/columns | Free-form text |
| **Analysis Speed** | Fast — direct aggregation | Slow — requires NLP processing |
| **Insights** | Sales totals, store rankings, category trends | Customer emotions, pain points, product quality |
| **Tools** | Pandas groupby, pivot tables | NLP, sentiment analysis, word clouds |
| **Challenges** | Data cleaning, missing values | Ambiguity, sarcasm, multiple languages |
| **Value** | Quantitative KPIs | Qualitative customer experience |

**Conclusion:** Structured data answers **WHAT** happened (sales dropped 15%). Unstructured data explains **WHY** (customers complained about fabric quality). Together, they provide a complete picture.

##  Exercise 5: Movie Ratings Dataset Analysis
**Dataset:** MovieLens ratings.csv  
**Goal:** Load, display, identify structured elements, and explain why they are structured.

```python
# Load MovieLens ratings dataset (simulated ratings.csv structure)
np.random.seed(7)
n = 300

df_ratings = pd.DataFrame({
    'userId':    np.random.randint(1, 500, n),
    'movieId':   np.random.randint(1, 1000, n),
    'rating':    np.random.choice([0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0], n),
    'timestamp': np.random.randint(964982703, 1537799250, n)
})

# Convert timestamp to readable datetime
df_ratings['datetime'] = pd.to_datetime(df_ratings['timestamp'], unit='s')

print('=== MovieLens Ratings Dataset: First Rows ===')
print(f'Shape: {df_ratings.shape[0]} rows x {df_ratings.shape[1]} columns')
print()
df_ratings.head(10)
```

```python
# Identify structured elements
print('=== Structured Elements in ratings.csv ===')
print()

structured_elements = pd.DataFrame({
    'Column': ['userId', 'movieId', 'rating', 'timestamp', 'datetime'],
    'Type': ['Structured', 'Structured', 'Structured', 'Structured', 'Structured'],
    'Data_Type': ['Discrete Integer', 'Discrete Integer', 'Ordinal Numeric',
                  'Unix Timestamp', 'DateTime'],
    'Why_Structured': [
        'Integer ID with fixed format — references a unique user in a lookup table',
        'Integer ID with fixed format — references a unique movie in a lookup table',
        'Predefined scale (0.5 to 5.0 in 0.5 steps) — discrete ordered numeric values',
        'Unix epoch integer — precise, standardized numeric time representation',
        'ISO datetime derived from timestamp — sortable, filterable, human-readable'
    ]
})

structured_elements
```

```python
# Statistical analysis of ratings
print('=== Rating Distribution ===')
print(df_ratings['rating'].value_counts().sort_index())
print()
print(f'Mean rating   : {df_ratings["rating"].mean():.2f}')
print(f'Median rating : {df_ratings["rating"].median():.1f}')
print(f'Std deviation : {df_ratings["rating"].std():.2f}')
print(f'Unique users  : {df_ratings["userId"].nunique()}')
print(f'Unique movies : {df_ratings["movieId"].nunique()}')
```

### Why ratings.csv is Fully Structured Data

All columns in `ratings.csv` are **structured** because:

1. **Fixed Schema:** Every row has exactly 4 fields (userId, movieId, rating, timestamp) — no missing or variable-length fields.
2. **Predefined Data Types:** Each column has a specific, consistent data type (integer, float, timestamp).
3. **Constrained Values:** Ratings are limited to a fixed scale (0.5–5.0) with no free-form input allowed.
4. **Machine-Readable:** All values can be directly processed by SQL, Pandas, or any analytical tool without preprocessing.
5. **Relational:** `userId` and `movieId` serve as foreign keys linkable to other structured tables (users.csv, movies.csv).

```python
# Visualization: Rating distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Rating distribution
rating_counts = df_ratings['rating'].value_counts().sort_index()
axes[0].bar(rating_counts.index, rating_counts.values,
            width=0.4, color='steelblue', edgecolor='white')
axes[0].set_title('Rating Distribution (MovieLens)')
axes[0].set_xlabel('Rating')
axes[0].set_ylabel('Number of Ratings')
axes[0].set_xticks([0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0])

# Ratings over time
df_ratings['year'] = df_ratings['datetime'].dt.year
ratings_by_year = df_ratings.groupby('year')['rating'].mean()
axes[1].plot(ratings_by_year.index, ratings_by_year.values,
             marker='o', color='steelblue', linewidth=2)
axes[1].set_title('Average Rating by Year')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Average Rating')

plt.tight_layout()
plt.show()
```

##  Exercise 6: Generate a Synthetic Product Catalogue
**Tool:** Faker library  
**Goal:** Generate 500 synthetic products with ID, name, description, and price.

```python
from faker import Faker
import random

fake = Faker()
Faker.seed(42)
random.seed(42)

print(' Faker imported and initialized successfully')
```

```python
# Product categories and adjectives for realistic names
categories = ['Electronics', 'Clothing', 'Books', 'Home & Garden', 'Sports',
              'Beauty', 'Toys', 'Food & Beverage', 'Automotive', 'Office Supplies']

adjectives = ['Premium', 'Deluxe', 'Classic', 'Ultra', 'Pro', 'Smart',
              'Eco', 'Vintage', 'Modern', 'Compact', 'Advanced', 'Essential']

product_nouns = ['Kit', 'Set', 'Pack', 'Bundle', 'Edition', 'Series',
                 'Collection', 'Box', 'Starter', 'Plus', 'Max', 'Lite']

def generate_product(product_id):
    """Generate a single synthetic product entry."""
    category = random.choice(categories)
    adj = random.choice(adjectives)
    noun = random.choice(product_nouns)

    # Unique product ID with prefix
    unique_id = f'PROD-{product_id:05d}'

    # Realistic product name
    name = f"{adj} {fake.word().capitalize()} {noun} — {category}"

    # Detailed product description
    description = (
        f"{fake.sentence(nb_words=8)} "
        f"{fake.sentence(nb_words=10)} "
        f"Perfect for {fake.word()} enthusiasts. "
        f"{fake.sentence(nb_words=7)}"
    )

    # Realistic price based on category
    price_ranges = {
        'Electronics': (29.99, 1299.99),
        'Clothing': (9.99, 299.99),
        'Books': (4.99, 49.99),
        'Home & Garden': (14.99, 499.99),
        'Sports': (19.99, 599.99),
        'Beauty': (5.99, 149.99),
        'Toys': (7.99, 199.99),
        'Food & Beverage': (2.99, 79.99),
        'Automotive': (9.99, 899.99),
        'Office Supplies': (3.99, 199.99)
    }
    low, high = price_ranges[category]
    price = round(random.uniform(low, high), 2)

    return {
        'Product_ID': unique_id,
        'Name': name,
        'Category': category,
        'Description': description,
        'Price': price,
        'Stock': random.randint(0, 500),
        'Rating': round(random.uniform(1.0, 5.0), 1),
        'Brand': fake.company(),
        'SKU': fake.bothify(text='??-####-??').upper()
    }

# Generate 500 products
products = [generate_product(i) for i in range(1, 501)]
df_catalogue = pd.DataFrame(products)

print('=== Synthetic Product Catalogue: First Rows ===')
print(f'Shape: {df_catalogue.shape[0]} products x {df_catalogue.shape[1]} columns')
print()
df_catalogue.head()
```

```python
# Catalogue statistics
print('=== Catalogue Statistics ===')
print(f'Total Products  : {len(df_catalogue)}')
print(f'Total Categories: {df_catalogue["Category"].nunique()}')
print(f'Price Range     : ${df_catalogue["Price"].min():.2f} — ${df_catalogue["Price"].max():.2f}')
print(f'Average Price   : ${df_catalogue["Price"].mean():.2f}')
print(f'Average Rating  : {df_catalogue["Rating"].mean():.2f}')
print()
print('=== Products per Category ===')
print(df_catalogue['Category'].value_counts())
```

```python
# Visualizations for the catalogue
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Average price by category
avg_price = df_catalogue.groupby('Category')['Price'].mean().sort_values(ascending=False)
colors = plt.cm.Set2(range(len(avg_price)))
axes[0].barh(avg_price.index, avg_price.values, color=colors)
axes[0].set_title('Average Price by Category')
axes[0].set_xlabel('Average Price ($)')
axes[0].set_ylabel('Category')

# Rating distribution
axes[1].hist(df_catalogue['Rating'], bins=10, color='steelblue', edgecolor='white')
axes[1].set_title('Product Rating Distribution')
axes[1].set_xlabel('Rating (1.0 - 5.0)')
axes[1].set_ylabel('Number of Products')

plt.tight_layout()
plt.show()
```

```python
# Export the catalogue
df_catalogue.to_csv('product_catalogue.csv', index=False)
df_catalogue.to_json('product_catalogue.json', orient='records', indent=2)

print(' Catalogue exported to product_catalogue.csv')
print(' Catalogue exported to product_catalogue.json')
print(f'\n Total products generated: {len(df_catalogue)}')
```
