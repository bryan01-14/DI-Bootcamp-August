##  Exercice 1 : Introduction à l'Analyse de Données

### Qu'est-ce que l'analyse de données ?

L'**analyse de données** est le processus consistant à collecter, nettoyer, transformer et modéliser des données brutes dans le but d'en extraire des informations utiles, de tirer des conclusions et de soutenir la prise de décision. Elle combine des méthodes statistiques, des outils informatiques et une expertise métier pour révéler des tendances, des patterns et des anomalies cachées dans les données.

---

### Pourquoi l'analyse des données est-elle importante aujourd'hui ?

Dans un monde où des milliards de données sont générées chaque seconde (transactions bancaires, clics web, capteurs IoT, réseaux sociaux), l'analyse de données est devenue un **avantage concurrentiel stratégique** :

1. **Prise de décision éclairée** : plutôt que de s'appuyer sur l'intuition, les organisations fondent leurs décisions sur des preuves empiriques.
2. **Optimisation des performances** : identification des goulots d'étranglement, des opportunités d'amélioration et des comportements anormaux.
3. **Personnalisation à grande échelle** : recommandation de produits, services adaptés et expériences utilisateur sur mesure.
4. **Anticipation des risques** : modèles prédictifs pour détecter fraudes, maladies ou pannes avant qu'elles ne surviennent.

---

### Trois domaines d'application de l'analyse de données

| Domaine | Application concrète |
|--------|----------------------|
| **Santé** | Analyse des dossiers médicaux pour prédire les rechutes, optimiser les traitements (ex : détection précoce de la drépanocytose par clustering de résultats de NFS) |
| **Finance** | Détection de fraudes par carte bancaire en temps réel grâce aux algorithmes de machine learning (ex : isolation forest, random forest) |
| **E-commerce & Marketing** | Segmentation client, analyse du parcours d'achat, optimisation des campagnes publicitaires (ex : A/B testing, analyse de cohortes) |

###  Dataset 1 — How Much Sleep Do Americans Really Get?


```python
# ── Dataset 1 : How Much Sleep Do Americans Really Get?
# Source : fichier ZIP fourni par le cours (GitHub devtlv)
import requests, zipfile, io

URL_SLEEP_ZIP = (
    "https://github.com/devtlv/Datasets-DA-Bootcamp-2-/raw/refs/heads/main/"
    "Week%204%20-%20Data%20Understanding/"
    "W4D2%20-%20Data%20U.%20and%20pre-processing/"
    "How%20Much%20Sleep%20Do%20Americans.zip"
)

response = requests.get(URL_SLEEP_ZIP)
response.raise_for_status()

with zipfile.ZipFile(io.BytesIO(response.content)) as zf:
    # Afficher les fichiers contenus dans le ZIP
    print("Fichiers dans le ZIP :", zf.namelist())
    # Charger le premier CSV trouvé
    csv_name = [n for n in zf.namelist() if n.endswith('.csv')][0]
    with zf.open(csv_name) as f:
        df_sleep = pd.read_csv(f)

print(f"\n Dataset Sleep chargé — {df_sleep.shape[0]} lignes × {df_sleep.shape[1]} colonnes")
print("\n── Premières lignes ──")
df_sleep.head()
```

## Exercice 6 : Données Structurées vs Non Structurées

### Classification des sources de données

| # | Source | Classification | Justification |
|---|--------|----------------|---------------|
| 1 | Rapports financiers d'une entreprise dans un fichier Excel | **Structurée** | Données organisées en lignes et colonnes avec un schéma fixe, directement requêtables |
| 2 | Photographies publiées sur un réseau social | **Non structurée** | Fichiers image sans schéma prédéfini ; le contenu sémantique ne peut être extrait qu'avec des algorithmes de vision par ordinateur |
| 3 | Collection d'articles de presse sur un site web | **Non structurée** | Texte libre sans format tabulaire ; nécessite du traitement NLP pour être exploité analytiquement |
| 4 | Données d'inventaire dans une base de données relationnelle | **Structurée** | Stockage en tables avec clés primaires, types définis et relations explicites |
| 5 | Entretiens enregistrés d'une étude de marché | **Non structurée** | Audio brut ; le contenu n'est pas directement analysable sans transcription et traitement du langage |

**Règle générale :**  
- *Structurée* = format tabulaire avec schéma défini, directement interrogeable par SQL ou Pandas.  
- *Non structurée* = texte libre, images, audio, vidéo — nécessite une transformation avant analyse.

##  Exercice 7 : Transformation de Données Non Structurées

### Méthodes de conversion en données structurées

---

#### 1. Série de billets de blog sur des expériences de voyage

**Méthode :** Extraction d'entités nommées (NER) + analyse de sentiment + structuration en DataFrame.

**Étapes :**
1. Scraper ou collecter les textes bruts des articles.
2. Appliquer un modèle NER (SpaCy, BERT) pour extraire : pays visité, date, activités mentionnées, évaluations.
3. Appliquer une analyse de sentiment (VADER, TextBlob) pour scorer chaque article.
4. Stocker dans un DataFrame : `[article_id, pays, date, score_sentiment, activités, auteur]`.

---

#### 2. Enregistrements audio de service client

**Méthode :** Speech-to-Text (STT) + classification de l'intention + analyse de sentiment.

**Étapes :**
1. Transcrire les audio via un service STT (Google Speech API, Whisper d'OpenAI).
2. Identifier les entités clés : numéro client, problème signalé, durée de l'appel.
3. Classifier l'intention (réclamation, demande d'info, résiliation…) via un modèle NLP.
4. Stocker dans un DataFrame : `[call_id, client_id, durée, problème, intention, résolution, satisfaction]`.

---

#### 3. Notes manuscrites d'une session de brainstorming

**Méthode :** OCR (reconnaissance optique de caractères) + extraction de mots-clés + clustering.

**Étapes :**
1. Scanner ou photographier les notes.
2. Appliquer un OCR (Tesseract, Google Vision) pour obtenir le texte.
3. Nettoyer le texte (correction orthographique) et extraire les mots-clés (TF-IDF).
4. Regrouper les idées par thème (clustering K-means ou LDA).
5. Stocker dans un DataFrame : `[idée_id, texte, thème, auteur_supposé, date]`.

---

#### 4. Tutoriel vidéo de cuisine

**Méthode :** Transcription audio + extraction d'images clés + structuration de la recette.

**Étapes :**
1. Extraire la piste audio et la transcrire (Whisper).
2. Extraire des images clés (OpenCV) aux moments où les ingrédients sont affichés.
3. Parser la transcription pour identifier : titre, ingrédients (quantités + unités), étapes de préparation, temps de cuisson.
4. Stocker dans un DataFrame structuré ou JSON : `[étape, durée_secondes, ingrédients, action]`.

##  Exercice 10 : Lecture de Données JSON depuis une URL


```python
# ── Exercice 10 : Lecture de données JSON depuis le ZIP du cours
# Source : posts.zip — dépôt GitHub devtlv

URL_POSTS_ZIP = (
    "https://github.com/devtlv/Datasets-DA-Bootcamp-2-/raw/refs/heads/main/"
    "Week%204%20-%20Data%20Understanding/"
    "W4D3%20-%20Importing%20Data%2C%20Exporting%20D/posts.zip"
)

response = requests.get(URL_POSTS_ZIP)
response.raise_for_status()

with zipfile.ZipFile(io.BytesIO(response.content)) as zf:
    print("Fichiers dans le ZIP :", zf.namelist())
    # Chercher le premier fichier JSON
    json_files = [n for n in zf.namelist() if n.endswith('.json')]
    if json_files:
        with zf.open(json_files[0]) as f:
            df_json = pd.read_json(f)
    else:
        # Si le ZIP contient un CSV, le charger comme fallback
        csv_files = [n for n in zf.namelist() if n.endswith('.csv')]
        with zf.open(csv_files[0]) as f:
            df_json = pd.read_csv(f)

print(f"\n Données JSON chargées — {df_json.shape[0]} lignes × {df_json.shape[1]} colonnes")
print("\n── Cinq premières entrées ──")
df_json.head()
```

```python
print("── Informations sur le dataset JSON ──")
df_json.info()
print()
print("── Noms des colonnes ──")
print(df_json.columns.tolist())
```

```python
# Exploration : affichage structuré des champs imbriqués (foods.likes, foods.dislikes)
print("Détail des préférences alimentaires par animal :")
print()
for _, row in df_json.iterrows():
    foods = row.get('foods', {})
    likes = foods.get('likes', []) if isinstance(foods, dict) else []
    dislikes = foods.get('dislikes', []) if isinstance(foods, dict) else []
    print(f"  🐾 {row['name']} ({row['species']})")
    print(f"      Aime    : {', '.join(likes)}")
    print(f"      N'aime pas : {', '.join(dislikes)}")
    print()
```
