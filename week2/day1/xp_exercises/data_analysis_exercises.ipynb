##  Exercise 1: Introduction to Data Analysis
### Essay / Report

### What is Data Analysis?
Data analysis is the process of **inspecting, cleaning, transforming, and modeling data** with the goal of discovering useful information, drawing conclusions, and supporting decision-making. It involves applying statistical and logical techniques to describe, summarize, and evaluate data in a systematic way.

---

### Why is Data Analysis Important in Modern Contexts?
In today's data-driven world, organizations generate massive volumes of data every second. Data analysis is critical because it:

- **Enables informed decision-making**: Instead of relying on intuition, businesses use data evidence to guide strategies.
- **Identifies trends and patterns**: Analysts can detect market shifts, customer behavior, and operational inefficiencies early.
- **Reduces risk**: Predictive analytics helps anticipate future outcomes, minimizing uncertainty.
- **Increases efficiency**: By analyzing workflows, companies optimize resource allocation and reduce waste.
- **Drives innovation**: Data insights reveal unmet customer needs and new market opportunities.

---

### Three Real-World Domains Where Data Analysis is Applied

#### 1.  Healthcare
Data analysis is used to study patient outcomes, predict disease outbreaks, personalize treatments, and optimize hospital resource management. For example, machine learning models analyze medical imaging data to detect cancer earlier than traditional methods.

#### 2. Finance & Banking
Banks use data analysis for fraud detection, credit risk assessment, and algorithmic trading. By analyzing transaction patterns in real time, financial institutions can flag suspicious activity and protect customers from fraud.

#### 3.  Retail & E-Commerce
Retailers like Amazon analyze customer purchase history, browsing behavior, and demographics to power recommendation engines, optimize pricing, and manage inventory. This increases customer satisfaction and maximizes revenue.

### Interesting Columns for Analysis

| Column | Analysis Type | Reasoning |
|---|---|---|
| `Hours_of_Sleep` | **Trend Analysis** | Track average sleep over age groups to detect if older people sleep less |
| `Gender` + `Hours_of_Sleep` | **Group Comparison** | Compare average sleep hours between males and females |
| `Work_Hours` + `Hours_of_Sleep` | **Correlation** | Analyze if people working more hours sleep less |
| `Sleep_Quality` | **Distribution** | Identify the most common sleep quality level across the population |
| `Exercise_Frequency` + `Sleep_Quality` | **Group Comparison** | Determine if people who exercise more report better sleep quality |

```python
# Visualization: Sleep hours distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Histogram of sleep hours
axes[0].hist(df_sleep['Hours_of_Sleep'], bins=6, color='steelblue', edgecolor='white')
axes[0].set_title('Distribution of Sleep Hours')
axes[0].set_xlabel('Hours of Sleep')
axes[0].set_ylabel('Frequency')

# Bar chart: Sleep quality count
quality_counts = df_sleep['Sleep_Quality'].value_counts()
axes[1].bar(quality_counts.index, quality_counts.values, color=['#e74c3c','#f39c12','#2ecc71','#3498db'])
axes[1].set_title('Sleep Quality Distribution')
axes[1].set_xlabel('Sleep Quality')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

##  Exercise 6: Structured vs Unstructured Data

### Classification: Structured or Unstructured?

| # | Data Source | Type | Reasoning |
|---|---|---|---|
| 1 | Financial reports stored in an Excel file | **Structured** | Excel files are organized in rows and columns with defined headers — a classic structured format |
| 2 | Photographs uploaded to a social media platform | **Unstructured** | Images have no inherent tabular organization; they are raw binary data |
| 3 | A collection of news articles on a website | **Unstructured** | Free-form text with no fixed schema — length, format, and content vary |
| 4 | Inventory data in a relational database | **Structured** | Relational databases store data in tables with defined columns and data types |
| 5 | Recorded interviews from a market research study | **Unstructured** | Audio recordings are raw media without a predefined schema or tabular format |

##  Exercise 7: Data Transformation Exercise
Converting unstructured data into structured formats.

### Proposed Transformation Methods

#### 1.  Travel Blog Articles
**Method:** Natural Language Processing (NLP) + Named Entity Recognition (NER)  
Extract structured fields such as: `Author`, `Date`, `Destination`, `Duration`, `Activities`, `Rating`, `Budget`.  
**Result:** A table where each row = one blog post with extracted attributes as columns.

---

#### 2.  Customer Service Call Recordings
**Method:** Automatic Speech Recognition (ASR) + Sentiment Analysis  
Transcribe audio to text, then extract: `Call_ID`, `Date`, `Duration`, `Customer_ID`, `Issue_Type`, `Sentiment_Score`, `Resolution_Status`.  
**Result:** A structured table allowing trend analysis and agent performance tracking.

---

#### 3.  Handwritten Brainstorming Notes
**Method:** Optical Character Recognition (OCR) + Text Categorization  
Use OCR to digitize text, then classify ideas into categories: `Idea_ID`, `Category`, `Priority`, `Author`, `Date`, `Description`.  
**Result:** A structured database of ideas that can be sorted, filtered, and prioritized.

---

#### 4.  Cooking Tutorial Videos
**Method:** Video Transcription + Computer Vision + NLP  
Transcribe spoken content, extract visual cues for ingredients, then structure: `Recipe_Name`, `Duration_Minutes`, `Ingredients`, `Steps_Count`, `Cuisine_Type`, `Difficulty_Level`, `Chef_Name`.  
**Result:** A recipe database that can be searched and filtered by cuisine, difficulty, or ingredient.

##  Exercise 9: Export DataFrame to Excel and JSON

```python
# Create a simple DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [28, 34, 22, 45, 31],
    'City': ['Paris', 'London', 'New York', 'Tokyo', 'Sydney'],
    'Salary': [55000, 72000, 45000, 98000, 63000],
    'Department': ['Engineering', 'Marketing', 'Design', 'Finance', 'Engineering']
}

df_simple = pd.DataFrame(data)
print('=== Simple DataFrame ===')
df_simple
```

```python
# Export to Excel
df_simple.to_excel('employees.xlsx', index=False)
print(' DataFrame exported to employees.xlsx')

# Verify by reading it back
df_from_excel = pd.read_excel('employees.xlsx')
print('\n=== Verified from Excel ===')
df_from_excel
```

```python
# Export to JSON
df_simple.to_json('employees.json', orient='records', indent=2)
print(' DataFrame exported to employees.json')

# Verify by reading it back
df_from_json = pd.read_json('employees.json')
print('\n=== Verified from JSON ===')
df_from_json
```
