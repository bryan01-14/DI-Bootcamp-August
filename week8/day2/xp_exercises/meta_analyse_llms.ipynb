# Méta-analyse d'articles de recherche sur les grands modèles de langage (LLM)

**Thème :** mise à l'échelle, architecture et alignement des LLM

*Mini-projet 1 — Méta-analyse d'articles de recherche en LLM*

## 1. Introduction

Les grands modèles de langage (Large Language Models, LLM) sont des réseaux de neurones de type transformeur, entraînés sur d'immenses corpus de texte afin de prédire la suite d'une séquence de mots. Depuis 2020, la taille de ces modèles et des données utilisées pour les entraîner a explosé, donnant naissance à des systèmes capables de rédiger, résumer, traduire, raisonner et dialoguer avec un niveau de compétence proche de l'humain dans de nombreuses tâches. Cette progression rapide soulève cependant des questions essentielles : comment allouer au mieux les ressources de calcul disponibles, quelles architectures permettent de gagner en efficacité sans sacrifier la qualité, et comment aligner le comportement de ces modèles sur les intentions et les valeurs humaines.

L'objectif de cette méta-analyse est de comparer cinq articles de recherche influents qui, ensemble, retracent une partie de cette évolution : le passage d'une course à la taille des modèles vers une recherche d'efficacité (mise à l'échelle optimale, architectures à mélange d'experts) et vers un alignement plus fin sur les préférences humaines (apprentissage par renforcement à partir de retours humains, instruction tuning). Le fil conducteur retenu est donc double : l'efficacité computationnelle des LLM (comment obtenir la meilleure performance pour un budget de calcul donné) et leur alignement (comment rendre leurs sorties utiles, sûres et fidèles aux instructions des utilisateurs). Ces deux axes sont profondément liés dans la littérature récente, puisqu'un modèle mieux entraîné dès la phase de pré-entraînement facilite ensuite un alignement plus efficace.

Les cinq articles retenus, présentés en détail dans la section suivante, sont :

1. Training Compute-Optimal Large Language Models (Chinchilla) — Hoffmann et al., 2022, NeurIPS / arXiv.
2. Training language models to follow instructions with human feedback (InstructGPT) — Ouyang et al., 2022, NeurIPS / arXiv.
3. Llama 2: Open Foundation and Fine-Tuned Chat Models — Touvron et al., 2023, Meta AI / arXiv.
4. Mixtral of Experts — Jiang et al., 2024, Mistral AI / arXiv.
5. GPT-4 Technical Report — OpenAI, 2023, arXiv.

## 2. Résumés des articles

### 2.1 Training Compute-Optimal Large Language Models (Chinchilla)

**Référence complète :** Hoffmann, J., Borgeaud, S., Mensch, A., Buchatskaya, E., Cai, T., Rutherford, E., et al. (2022). *Training Compute-Optimal Large Language Models*. Advances in Neural Information Processing Systems (NeurIPS 2022) ; arXiv:2203.15556.

**Problème de recherche :** au début des années 2020, la tendance dominante consistait à augmenter la taille des modèles (nombre de paramètres) tout en gardant à peu près constante la quantité de données d'entraînement. Les auteurs, chez DeepMind, se demandent si cette stratégie utilise réellement de façon optimale le budget de calcul disponible.

**Solution proposée :** les auteurs entraînent plus de 400 modèles transformeurs, de 70 millions à plus de 16 milliards de paramètres, sur des jeux de données allant de 5 à 500 milliards de jetons (tokens). Grâce à trois approches empiriques complémentaires, ils établissent une loi de mise à l'échelle selon laquelle, pour un budget de calcul fixé, la taille du modèle et le nombre de jetons d'entraînement doivent croître à un rythme comparable : chaque doublement de la taille du modèle devrait s'accompagner d'un doublement du volume de données. Ils valident cette hypothèse en entraînant Chinchilla, un modèle de 70 milliards de paramètres entraîné sur environ 1,4 trillion de jetons, avec le même budget de calcul que Gopher (280 milliards de paramètres, entraîné sur environ 300 milliards de jetons).

**Principaux résultats :** Chinchilla surpasse uniformément des modèles beaucoup plus grands comme Gopher, GPT-3 et Megatron-Turing NLG sur un large éventail de tâches d'évaluation, tout en étant nettement moins coûteux à utiliser en inférence et en réglage fin. Il atteint notamment une précision moyenne état de l'art de 67,5 % sur le benchmark MMLU, soit une amélioration de plus de 7 points par rapport à Gopher.

**Architecture :** transformeur dense standard. **Données :** corpus massifs de texte en anglais (type MassiveText). **Métriques :** perplexité, précision sur MMLU, BIG-bench et diverses tâches de question-réponse.

### 2.2 Training Language Models to Follow Instructions with Human Feedback (InstructGPT)

**Référence complète :** Ouyang, L., Wu, J., Jiang, X., Almeida, D., Wainwright, C., Mishkin, P., et al. (2022). *Training Language Models to Follow Instructions with Human Feedback*. Advances in Neural Information Processing Systems (NeurIPS 2022) ; arXiv:2203.02155.

**Problème de recherche :** les modèles de langage comme GPT-3 sont entraînés à prédire le mot suivant, un objectif qui ne correspond pas nécessairement à ce que les utilisateurs souhaitent réellement (des réponses utiles, honnêtes et non nuisibles). Un modèle plus gros n'est pas automatiquement mieux aligné sur les intentions humaines.

**Solution proposée :** les auteurs, chez OpenAI, développent un pipeline en trois étapes d'apprentissage par renforcement à partir de retours humains (RLHF). D'abord, un modèle GPT-3 pré-entraîné est affiné par apprentissage supervisé sur des démonstrations rédigées par des annotateurs humains. Ensuite, des annotateurs classent plusieurs réponses du modèle par ordre de préférence, ce qui permet d'entraîner un modèle de récompense (reward model) qui apprend à prédire les préférences humaines. Enfin, la politique du modèle est optimisée pour maximiser cette récompense via l'algorithme Proximal Policy Optimization (PPO). Les modèles résultants sont appelés InstructGPT et sont déclinés en trois tailles : 1,3, 6 et 175 milliards de paramètres, tous basés sur l'architecture GPT-3.

**Principaux résultats :** les annotateurs préfèrent nettement les réponses d'InstructGPT (même la version à 1,3 milliard de paramètres) à celles de GPT-3 à 175 milliards de paramètres, malgré un facteur de taille supérieur à 100. InstructGPT montre également des gains en véracité et une réduction de la génération de contenus toxiques, avec une dégradation limitée des performances sur les benchmarks de traitement du langage naturel classiques.

**Données :** démonstrations et comparaisons humaines collectées sur des invites (prompts) réelles issues de l'API OpenAI. **Métriques :** préférence humaine (taux de victoire), véracité (TruthfulQA), toxicité (RealToxicityPrompts).

### 2.3 Llama 2: Open Foundation and Fine-Tuned Chat Models

**Référence complète :** Touvron, H., Martin, L., Stone, K., Albert, P., Almahairi, A., Babaei, Y., et al. (2023). *Llama 2: Open Foundation and Fine-Tuned Chat Models*. Meta AI ; arXiv:2307.09288.

**Problème de recherche :** les modèles conversationnels alignés les plus performants (ChatGPT, Bard, Claude) restent fermés, et le coût en calcul et en annotation humaine du processus d'alignement est rarement documenté de façon transparente, ce qui freine la recherche ouverte sur l'alignement.

**Solution proposée :** Meta publie Llama 2, une famille de modèles de fondation pré-entraînés (7, 13, 34 et 70 milliards de paramètres) ainsi que leur version affinée pour le dialogue, Llama 2-Chat. Les modèles reprennent une architecture transformeur dense avec attention par requêtes groupées (Grouped-Query Attention) pour les plus grandes tailles, afin de réduire le coût mémoire de l'inférence. L'alignement conversationnel combine un réglage fin supervisé, un RLHF avec deux modèles de récompense distincts (un pour l'utilité, un pour la sécurité), une méthode de rééchantillonnage par rejet (rejection sampling) en complément de PPO, ainsi qu'une technique nommée Ghost Attention conçue pour maintenir la cohérence des instructions au fil d'une conversation à plusieurs tours.

**Principaux résultats :** Llama 2-Chat obtient des résultats jugés comparables à certains modèles propriétaires sur des évaluations humaines d'utilité et de sécurité, tout en étant publié en open source (licence communautaire Meta), ce qui en a fait une base largement réutilisée par la communauté.

**Données :** corpus public retravaillé (2 trillions de jetons pour le pré-entraînement) et données de préférence humaine collectées en interne. **Métriques :** évaluations humaines comparatives, benchmarks académiques standards (raisonnement de bon sens, connaissances, code), tests de sécurité par équipes rouges (red teaming).

### 2.4 Mixtral of Experts

**Référence complète :** Jiang, A. Q., Sablayrolles, A., Roux, A., Mensch, A., Savary, B., Bamford, C., et al. (2024). *Mixtral of Experts*. Mistral AI ; arXiv:2401.04088.

**Problème de recherche :** les modèles denses activent l'intégralité de leurs paramètres pour chaque jeton traité, ce qui lie strictement la capacité du modèle au coût de calcul en inférence. La question posée est de savoir comment augmenter la capacité effective d'un modèle sans faire croître proportionnellement son coût d'utilisation.

**Solution proposée :** Mixtral 8x7B reprend l'architecture de Mistral 7B, mais chaque couche du réseau de propagation avant (feed-forward) est remplacée par huit blocs experts. Pour chaque jeton, un réseau routeur sélectionne dynamiquement deux experts parmi les huit afin de traiter l'état courant et combine leurs sorties : c'est une architecture de type Sparse Mixture of Experts (SMoE). Le modèle possède au total 46,7 milliards de paramètres, mais n'en active qu'environ 12,9 milliards par jeton, ce qui réduit fortement le coût d'inférence par rapport à un modèle dense de taille équivalente. Le contexte d'entraînement s'étend jusqu'à 32 000 jetons.

**Principaux résultats :** Mixtral égale ou surpasse Llama 2 70B et GPT-3.5 sur l'ensemble des benchmarks évalués, tout en n'activant qu'une fraction de leurs paramètres, avec une inférence environ six fois plus rapide que Llama 2 70B à qualité comparable. Il se distingue en particulier en mathématiques, en génération de code et dans les tâches multilingues. Sa version affinée pour le dialogue (Mixtral 8x7B – Instruct) dépasse GPT-3.5 Turbo, Claude 2.1, Gemini Pro et Llama 2 70B-Chat sur des évaluations humaines. Les deux modèles sont publiés sous licence Apache 2.0.

**Données :** corpus multilingue non détaillé publiquement. **Métriques :** benchmarks académiques standards (MMLU, raisonnement, mathématiques, code), évaluations humaines pour la version Instruct.

### 2.5 GPT-4 Technical Report

**Référence complète :** OpenAI (2023). *GPT-4 Technical Report*. arXiv:2303.08774.

**Problème de recherche :** à mesure que les modèles grandissent, il devient plus difficile de prévoir leurs performances avant la fin d'un entraînement extrêmement coûteux ; les auteurs cherchent également à étendre les capacités des LLM au-delà du texte, vers la compréhension multimodale, tout en maîtrisant les risques associés à des modèles plus puissants.

**Solution proposée :** GPT-4 est un modèle transformeur multimodal capable d'accepter des entrées combinant texte et image et de produire des sorties textuelles. Une contribution méthodologique majeure du rapport est le développement d'une infrastructure d'entraînement et de méthodes d'optimisation permettant de prédire, à partir de modèles beaucoup plus petits (jusqu'à 1000 fois moins de calcul), certains aspects des performances finales de GPT-4 avant même de lancer l'entraînement complet. Comme pour InstructGPT, un processus d'alignement post-entraînement (RLHF) est appliqué afin d'améliorer la factualité et le respect du comportement souhaité. Par choix de transparence limitée sur la sécurité et la compétitivité, OpenAI ne divulgue pas les détails de l'architecture, de la taille du modèle ni des données d'entraînement.

**Principaux résultats :** GPT-4 atteint un niveau de performance comparable à l'humain sur de nombreux benchmarks professionnels et académiques, par exemple en réussissant un examen du barreau simulé avec un score situé dans le top 10 % des candidats, et en obtenant environ 86,4 % sur MMLU. Le processus d'alignement réduit également la toxicité et améliore l'adhésion aux politiques d'usage par rapport aux versions précédentes de GPT.

**Données et architecture :** non divulguées. **Métriques :** examens professionnels et académiques simulés, MMLU, HumanEval (code), évaluations de sécurité internes.

## 3. Analyse comparative

Le tableau ci-dessous synthétise les objectifs, choix architecturaux, stratégies d'entraînement et résultats d'évaluation des cinq articles.

| Article | Objectif principal | Architecture | Stratégie d'entraînement | Benchmarks clés |
| --- | --- | --- | --- | --- |
| Chinchilla (2022) | Optimiser l'allocation calcul/données | Transformeur dense (70B) | Pré-entraînement seul, loi de mise à l'échelle | MMLU, BIG-bench |
| InstructGPT (2022) | Aligner les sorties sur l'intention humaine | GPT-3 dense (1,3B–175B) | SFT + reward model + PPO (RLHF) | Préférence humaine, TruthfulQA |
| Llama 2 (2023) | Modèle ouvert compétitif et aligné | Transformeur dense + GQA (7B–70B) | Pré-entraînement + SFT + RLHF double (utilité/sécurité) | Évaluations humaines, benchmarks académiques |
| Mixtral (2024) | Capacité élevée à coût d'inférence réduit | Sparse Mixture-of-Experts (8x7B) | Pré-entraînement dense-like + routage top-2 | MMLU, code, mathématiques, multilingue |
| GPT-4 (2023) | Capacités multimodales prévisibles à grande échelle | Transformeur multimodal (non divulgué) | Pré-entraînement + RLHF, prédiction de performance | Examens professionnels, MMLU, HumanEval |

### 3.1 Objectifs et domaines de problème

Deux familles d'objectifs se dégagent. Chinchilla et Mixtral s'attaquent avant tout à l'efficacité computationnelle : le premier au niveau du pré-entraînement (combien de données pour combien de paramètres), le second au niveau de l'architecture d'inférence (comment activer moins de paramètres sans perdre en qualité). InstructGPT, Llama 2 et GPT-4 se concentrent davantage sur l'alignement et l'utilisabilité, c'est-à-dire la transformation d'un modèle de langage brut en assistant capable de suivre des instructions de façon utile et sûre. Llama 2 et GPT-4 combinent en réalité les deux préoccupations, en s'appuyant sur des choix d'efficacité (GQA, prédictibilité de la mise à l'échelle) en amont de leur processus d'alignement.

### 3.2 Architectures et innovations

Chinchilla, InstructGPT et GPT-4 (dans la mesure où l'information est publique) reposent sur des transformeurs denses classiques. Llama 2 introduit l'attention par requêtes groupées pour réduire le coût mémoire du cache d'attention en inférence. Mixtral se distingue nettement en remplaçant les couches denses par des couches à mélange d'experts activées de façon sparse, ce qui représente le changement architectural le plus significatif parmi les cinq articles : il découple pour la première fois dans ce corpus la taille totale du modèle de son coût d'inférence effectif.

### 3.3 Stratégies d'entraînement et de réglage fin

On observe une convergence progressive vers un pipeline en plusieurs étapes : pré-entraînement à grande échelle, puis réglage fin supervisé, puis alignement par retour humain. InstructGPT est le premier des cinq articles à formaliser ce pipeline RLHF complet (SFT, reward model, PPO). Llama 2 le reprend et l'enrichit avec deux modèles de récompense séparés et une méthode de rééchantillonnage par rejet en complément de PPO. GPT-4 s'appuie sur un RLHF similaire mais met l'accent sur la prédictibilité des résultats à partir de modèles plus petits. Chinchilla et Mixtral, en comparaison, se concentrent surtout sur le pré-entraînement et l'architecture plutôt que sur l'alignement post-entraînement (bien qu'une version Instruct de Mixtral existe).

### 3.4 Benchmarks et évaluation

MMLU s'impose comme le benchmark commun le plus cité, utilisé par Chinchilla, Llama 2, Mixtral et GPT-4, ce qui en fait un point de comparaison privilégié entre modèles de générations différentes. InstructGPT et Llama 2 accordent une place centrale aux évaluations humaines comparatives (préférence, sécurité), plus coûteuses à mettre en œuvre mais jugées plus représentatives de l'utilité réelle perçue par les utilisateurs que les benchmarks académiques automatisés seuls.

### 3.5 Forces, limites et reproductibilité

Chinchilla et Mixtral fournissent des détails méthodologiques relativement complets et, pour Mixtral, un modèle publié en licence ouverte (Apache 2.0), ce qui facilite la reproductibilité. Llama 2 va plus loin en documentant explicitement son processus RLHF et en publiant les poids du modèle sous licence communautaire, tout en restant plus fermé que Mixtral sur certains détails de données. InstructGPT est bien documenté sur le plan méthodologique mais s'appuie sur des données d'annotation humaine propriétaires, difficiles à reproduire à l'identique. GPT-4 constitue le cas le moins reproductible du corpus : ni l'architecture, ni la taille, ni les données d'entraînement ne sont divulguées, ce qui limite fortement la vérification indépendante des résultats malgré leur solidité apparente.

## 4. Insights et réflexion

### 4.1 Tendances générales

Le corpus étudié illustre un déplacement progressif du critère de succès des LLM : de « plus gros est meilleur » vers « mieux entraîné et mieux aligné est meilleur ». Chinchilla démontre dès 2022 que de nombreux modèles étaient sous-entraînés au regard de leur taille ; Mixtral prolonge cette logique d'efficacité côté inférence. En parallèle, InstructGPT, Llama 2 et GPT-4 montrent qu'un alignement soigné peut rapprocher un modèle relativement modeste (InstructGPT 1,3B) du niveau perçu de modèles bien plus grands non alignés, ce qui déplace une partie de la valeur ajoutée de la seule échelle vers la qualité du post-entraînement.

### 4.2 Méthodes les plus prometteuses

Le RLHF, introduit et formalisé par InstructGPT, apparaît comme la méthode la plus structurante : elle est reprise et adaptée par Llama 2 et GPT-4, avec des variantes (double modèle de récompense, rééchantillonnage par rejet). Du côté architectural, le mélange d'experts illustré par Mixtral semble être une direction particulièrement prometteuse pour concilier capacité et coût d'inférence, une piste depuis reprise par plusieurs familles de modèles ouverts et fermés.

### 4.3 Limites communément reconnues

Plusieurs limites reviennent d'un article à l'autre : le coût et la subjectivité des annotations humaines nécessaires au RLHF (InstructGPT, Llama 2) ; le risque de biais et de sécurité qui subsiste malgré l'alignement, nécessitant des tests par équipes rouges (Llama 2, GPT-4) ; le manque de transparence sur les données et l'architecture pour les modèles les plus fermés (GPT-4), qui freine la recherche indépendante ; et, pour les architectures à mélange d'experts, une tendance du routeur à sur-solliciter certains experts, source de déséquilibres d'entraînement (Mixtral).

### 4.4 Orientations futures possibles

Les articles suggèrent plusieurs directions futures : l'exploration de lois de mise à l'échelle combinant à la fois données, paramètres et architectures sparses ; le développement de méthodes d'alignement moins coûteuses en annotation humaine, par exemple par retour automatisé ou par des modèles de récompense plus robustes ; l'amélioration de la stabilité du routage dans les architectures à mélange d'experts ; et une transparence accrue sur les données et les méthodes d'évaluation, afin de renforcer la reproductibilité scientifique face à la fermeture croissante de certains modèles de pointe.

## 5. Conclusion

Cette méta-analyse de cinq articles de recherche marquants sur les LLM met en évidence deux dynamiques complémentaires qui structurent le domaine depuis 2022 : la recherche d'une efficacité computationnelle accrue (Chinchilla, Mixtral) et le perfectionnement des méthodes d'alignement sur les intentions humaines (InstructGPT, Llama 2, GPT-4). Ces deux axes ne sont pas indépendants : un modèle mieux entraîné et architecturé plus efficacement (Chinchilla, Mixtral) constitue une meilleure base pour un alignement ultérieur, tandis qu'un alignement réussi (RLHF) permet de tirer davantage de valeur perçue d'un modèle sans nécessairement en augmenter la taille. Le domaine évolue ainsi vers des systèmes qui ne sont plus seulement évalués sur leur taille brute, mais sur le rapport entre performance, coût de calcul et fidélité aux attentes humaines. Les tensions persistantes autour de la reproductibilité, de la transparence et du coût de l'alignement humain restent toutefois des défis ouverts pour la recherche à venir.
