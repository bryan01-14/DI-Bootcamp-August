## Exercice 2 — Graphique Lineaire : Variation de Temperature

**Objectif** : visualiser les variations de temperature sur une semaine avec Matplotlib.


```python
# Donnees de temperature
days         = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']
temperatures = [72, 74, 76, 80, 82, 78, 75]

fig, ax = plt.subplots(figsize=(10, 5))
fig.patch.set_facecolor('#0f1117')

# Ligne principale
ax.plot(days, temperatures, color='#4FC3F7', linewidth=2.5,
        marker='o', markersize=9, markerfacecolor='#FF8A65',
        markeredgecolor='white', markeredgewidth=1.5, label='Temperature')

# Zone remplie sous la courbe
ax.fill_between(range(len(days)), temperatures, min(temperatures)-2,
                alpha=0.15, color='#4FC3F7')

# Annotations sur chaque point
for i, (day, temp) in enumerate(zip(days, temperatures)):
    ax.annotate(f'{temp}°F',
                xy=(i, temp),
                xytext=(0, 12),
                textcoords='offset points',
                ha='center', fontsize=10,
                color='#FFD54F', fontweight='bold')

# Ligne de moyenne
mean_temp = np.mean(temperatures)
ax.axhline(mean_temp, color='#81C784', linewidth=1.5,
           linestyle='--', alpha=0.8, label=f'Moyenne = {mean_temp:.1f}°F')

# Mise en forme
ax.set_xlabel('Jour', fontsize=12, labelpad=10)
ax.set_ylabel('Temperature (°F)', fontsize=12, labelpad=10)
ax.set_title('Variation de Temperature sur une Semaine', fontsize=14, fontweight='bold', pad=15)
ax.legend(fontsize=10)
ax.set_ylim(68, 87)
ax.grid(True, alpha=0.4, linestyle='--')

plt.tight_layout()
plt.show()

print(f"Temperature minimale : {min(temperatures)}°F ({days[temperatures.index(min(temperatures))]})")
print(f"Temperature maximale : {max(temperatures)}°F ({days[temperatures.index(max(temperatures))]})")
print(f"Temperature moyenne  : {mean_temp:.1f}°F")
```

## Exercice 3 — Graphique a Barres : Ventes Mensuelles

**Objectif** : visualiser les ventes mensuelles d'un magasin de detail avec Matplotlib.


```python
# Donnees de ventes mensuelles
months = ['Janvier', 'Fevrier', 'Mars', 'Avril', 'Mai']
sales  = [5000, 5500, 6200, 7000, 7500]

fig, ax = plt.subplots(figsize=(10, 5))
fig.patch.set_facecolor('#0f1117')

# Couleurs degradees du plus fonce au plus clair
colors = ['#1E88E5', '#26C6DA', '#42A5F5', '#64B5F6', '#4FC3F7']

bars = ax.bar(months, sales, color=colors, edgecolor='white',
              linewidth=0.8, alpha=0.9, width=0.6)

# Valeurs sur les barres
for bar, val in zip(bars, sales):
    ax.text(bar.get_x() + bar.get_width() / 2,
            bar.get_height() + 80,
            f'${val:,}',
            ha='center', va='bottom',
            fontsize=11, fontweight='bold', color='#FFD54F')

# Ligne de tendance
from scipy.stats import linregress
x_num = np.arange(len(months))
slope, intercept, *_ = linregress(x_num, sales)
x_line = np.linspace(-0.4, len(months)-0.6, 100)
ax.plot(x_line, slope * np.linspace(0, len(months)-1, 100) + intercept,
        color='#FF8A65', linewidth=2, linestyle='--', alpha=0.8, label=f'Tendance (+{slope:.0f}$/mois)')

# Mise en forme
ax.set_xlabel('Mois', fontsize=12, labelpad=10)
ax.set_ylabel('Montant des Ventes ($)', fontsize=12, labelpad=10)
ax.set_title('Ventes Mensuelles — Magasin de Detail', fontsize=14, fontweight='bold', pad=15)
ax.yaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'${x:,.0f}'))
ax.set_ylim(0, max(sales) * 1.18)
ax.legend(fontsize=10)
ax.grid(True, alpha=0.4, linestyle='--', axis='y')

plt.tight_layout()
plt.show()

growth = (sales[-1] - sales[0]) / sales[0] * 100
print(f"Croissance totale sur la periode : +{growth:.1f}%")
print(f"Croissance moyenne mensuelle     : +{slope:.0f}$ / mois")
```

## Exercice 4 — Histogramme : Distribution du CGPA

**Objectif** : visualiser la distribution des categories de CGPA des etudiants
avec `seaborn.histplot`.

Le CGPA (Cumulative Grade Point Average) est l'indicateur de performance academique.


```python
# Ordre logique des categories CGPA (du plus faible au plus eleve)
cgpa_order = ['0 - 1.99', '2.00 - 2.49', '2.50 - 2.99', '3.00 - 3.49', '3.50 - 4.00']

# Encoder en categoriel ordonne pour un tri correct
df['CGPA_cat'] = pd.Categorical(df['What is your CGPA?'],
                                categories=cgpa_order, ordered=True)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.patch.set_facecolor('#0f1117')

# --- Histogramme principal (seaborn histplot) ---
ax = axes[0]
sns.histplot(
    data=df,
    x='CGPA_cat',
    color='#4FC3F7',
    edgecolor='white',
    linewidth=0.8,
    alpha=0.85,
    ax=ax,
    shrink=0.8,
)
ax.set_title('Distribution du CGPA des Etudiants', fontsize=13, fontweight='bold', pad=12)
ax.set_xlabel('Tranche de CGPA', fontsize=11, labelpad=8)
ax.set_ylabel('Nombre d\'etudiants', fontsize=11, labelpad=8)
ax.grid(True, alpha=0.4, linestyle='--', axis='y')

# Annotations
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{int(height)}',
                    (p.get_x() + p.get_width() / 2., height),
                    xytext=(0, 6), textcoords='offset points',
                    ha='center', fontsize=10, color='#FFD54F', fontweight='bold')

# --- Pie chart complementaire ---
ax2 = axes[1]
cgpa_counts = df['CGPA_cat'].value_counts().reindex(cgpa_order)
colors_pie  = ['#FF8A65','#FFD54F','#81C784','#4FC3F7','#CE93D8']
wedges, texts, autotexts = ax2.pie(
    cgpa_counts.values,
    labels=cgpa_order,
    colors=colors_pie,
    autopct='%1.1f%%',
    startangle=90,
    pctdistance=0.82,
    textprops={'color': '#e0e0e0', 'fontsize': 9}
)
for at in autotexts: at.set_fontsize(8)
ax2.set_title('Repartition du CGPA (%)', fontsize=13, fontweight='bold', pad=12)

plt.suptitle('Exercice 4 — Distribution du CGPA (Student Mental Health)',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

print("Repartition du CGPA :")
for cat, cnt in zip(cgpa_order, cgpa_counts):
    print(f"  {cat} : {cnt} etudiants ({cnt/len(df)*100:.1f}%)")
```

## Exercice 5 — Graphique a Barres : Niveaux d'Anxiete par Genre

**Objectif** : comparer la proportion d'etudiants souffrant d'anxiete
selon le genre avec `seaborn.barplot`.


```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.patch.set_facecolor('#0f1117')

# --- 1. Barplot : proportion d'anxiete par genre ---
ax = axes[0]
anxiety_gender = (df.groupby(['Choose your gender','Do you have Anxiety?'])
                    .size()
                    .reset_index(name='count'))
anxiety_gender['proportion'] = (anxiety_gender
                                 .groupby('Choose your gender')['count']
                                 .transform(lambda x: x / x.sum()) * 100)
anxiety_yes = anxiety_gender[anxiety_gender['Do you have Anxiety?'] == 'Yes']

sns.barplot(
    data=anxiety_yes,
    x='Choose your gender',
    y='proportion',
    hue='Choose your gender',
    palette={'Female': '#F48FB1', 'Male': '#4FC3F7'},
    edgecolor='white',
    linewidth=0.8,
    alpha=0.88,
    ax=ax,
    legend=False,
)
ax.set_title('Proportion d\'Anxiete par Genre', fontsize=12, fontweight='bold', pad=10)
ax.set_xlabel('Genre', fontsize=11, labelpad=8)
ax.set_ylabel('Etudiants avec anxiete (%)', fontsize=11, labelpad=8)
ax.set_ylim(0, 100)
ax.grid(True, alpha=0.4, linestyle='--', axis='y')
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%',
                (p.get_x() + p.get_width()/2., p.get_height()),
                xytext=(0,8), textcoords='offset points',
                ha='center', fontsize=12, fontweight='bold', color='#FFD54F')

# --- 2. Barplot groupe : Oui/Non par genre ---
ax2 = axes[1]
sns.barplot(
    data=anxiety_gender,
    x='Choose your gender',
    y='proportion',
    hue='Do you have Anxiety?',
    palette={'Yes': '#FF8A65', 'No': '#81C784'},
    edgecolor='white',
    linewidth=0.8,
    alpha=0.88,
    ax=ax2,
)
ax2.set_title('Anxiete Oui/Non par Genre', fontsize=12, fontweight='bold', pad=10)
ax2.set_xlabel('Genre', fontsize=11, labelpad=8)
ax2.set_ylabel('Proportion (%)', fontsize=11, labelpad=8)
ax2.legend(title='Anxiete', fontsize=9)
ax2.grid(True, alpha=0.4, linestyle='--', axis='y')

# --- 3. Comparaison depression, anxiete, panique par genre ---
ax3 = axes[2]
conditions = ['Do you have Depression?', 'Do you have Anxiety?', 'Do you have Panic attack?']
cond_labels = ['Depression', 'Anxiete', 'Crises panique']
x  = np.arange(len(cond_labels))
w  = 0.3
for i, gender in enumerate(['Female','Male']):
    vals = [df[df['Choose your gender']==gender][c].eq('Yes').mean()*100 for c in conditions]
    color = '#F48FB1' if gender=='Female' else '#4FC3F7'
    bars = ax3.bar(x + (i-0.5)*w, vals, w, label=gender, color=color,
                   edgecolor='white', linewidth=0.8, alpha=0.88)
ax3.set_xticks(x); ax3.set_xticklabels(cond_labels, fontsize=10)
ax3.set_title('Sante Mentale par Genre (%)', fontsize=12, fontweight='bold', pad=10)
ax3.set_ylabel('Proportion (%)')
ax3.legend(fontsize=9); ax3.grid(True, alpha=0.4, axis='y', linestyle='--')

plt.suptitle('Exercice 5 — Anxiete et Sante Mentale par Genre',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Resume textuel
for gender in ['Female','Male']:
    pct = df[df['Choose your gender']==gender]['Do you have Anxiety?'].eq('Yes').mean()*100
    n_g = (df['Choose your gender']==gender).sum()
    print(f"{gender:<8} : {pct:.1f}% d'anxiete  (n={n_g})")
```

## Exercice 6 — Nuage de Points : Age et Crises de Panique

**Objectif** : explorer la relation entre l'age des etudiants et la survenue
de crises de panique avec `seaborn.scatterplot`.

Les reponses binaires (Oui/Non) sont converties en valeurs numeriques (1/0).


```python
# Conversion Oui/Non -> 1/0
df['Panic_numeric']    = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})
df['Anxiety_numeric']  = df['Do you have Anxiety?'].map({'Yes': 1, 'No': 0})
df['Depress_numeric']  = df['Do you have Depression?'].map({'Yes': 1, 'No': 0})

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.patch.set_facecolor('#0f1117')

# --- 1. Scatter : Age vs Crises de panique (colore par genre) ---
ax = axes[0]
jitter = np.random.uniform(-0.08, 0.08, len(df))  # eviter superposition

for gender, color, marker in [('Female','#F48FB1','o'), ('Male','#4FC3F7','s')]:
    mask = df['Choose your gender'] == gender
    sub  = df[mask]
    jit  = jitter[mask]
    ax.scatter(
        sub['Age'],
        sub['Panic_numeric'] + jit,
        c=color, marker=marker, s=70, alpha=0.65,
        edgecolors='white', linewidths=0.5, label=gender
    )

ax.set_yticks([0, 1])
ax.set_yticklabels(['Non (0)', 'Oui (1)'], fontsize=10)
ax.set_title('Age vs Crises de Panique', fontsize=12, fontweight='bold', pad=10)
ax.set_xlabel('Age', fontsize=11, labelpad=8)
ax.set_ylabel('Crise de panique ?', fontsize=11, labelpad=8)
ax.legend(title='Genre', fontsize=9)
ax.grid(True, alpha=0.4, linestyle='--')

# Pourcentage par age
age_panic = df.groupby('Age')['Panic_numeric'].mean() * 100
for age, pct in age_panic.items():
    ax.annotate(f'{pct:.0f}%',
                (age, 1.12),
                ha='center', fontsize=8, color='#FFD54F')
ax.text(df['Age'].min()-0.2, 1.12, 'Pct/age:', fontsize=7, color='#FFD54F', ha='right')

# --- 2. Scatter : Age vs Anxiete avec regression logistique visualisee ---
ax2 = axes[1]
sns.regplot(
    data=df, x='Age', y='Anxiety_numeric',
    scatter_kws={'color':'#CE93D8','alpha':0.55,'s':60,'edgecolors':'white','linewidths':0.5},
    line_kws={'color':'#FFD54F','lw':2.5,'linestyle':'--'},
    logistic=True, ci=95,
    ax=ax2
)
ax2.set_yticks([0, 1])
ax2.set_yticklabels(['Non (0)', 'Oui (1)'], fontsize=10)
ax2.set_title('Age vs Anxiete (regression logistique)', fontsize=12, fontweight='bold', pad=10)
ax2.set_xlabel('Age', fontsize=11, labelpad=8)
ax2.set_ylabel('Anxiete ?', fontsize=11, labelpad=8)
ax2.grid(True, alpha=0.4, linestyle='--')

# --- 3. Strip plot : CGPA vs crises de panique ---
ax3 = axes[2]
cgpa_order = ['0 - 1.99','2.00 - 2.49','2.50 - 2.99','3.00 - 3.49','3.50 - 4.00']
sns.stripplot(
    data=df,
    x='What is your CGPA?',
    y='Age',
    hue='Do you have Panic attack?',
    order=cgpa_order,
    palette={'Yes': '#FF8A65', 'No': '#81C784'},
    jitter=True,
    size=6,
    alpha=0.75,
    edgecolor='white',
    linewidth=0.4,
    ax=ax3
)
ax3.set_title('CGPA vs Age (crises de panique)', fontsize=12, fontweight='bold', pad=10)
ax3.set_xlabel('CGPA', fontsize=11, labelpad=8)
ax3.set_ylabel('Age', fontsize=11, labelpad=8)
ax3.tick_params(axis='x', rotation=30)
ax3.legend(title='Crise panique', fontsize=9)
ax3.grid(True, alpha=0.4, linestyle='--', axis='y')

plt.suptitle('Exercice 6 — Relations entre Age, CGPA et Sante Mentale',
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Statistiques
print("Taux de crises de panique par age :")
print(df.groupby('Age')['Panic_numeric'].agg(['mean','count'])
        .rename(columns={'mean':'Taux','count':'N'})
        .assign(Taux=lambda x: (x['Taux']*100).round(1))
        .to_string())
print(f"\nCorrelation Age ~ Crises de panique : r = {df['Age'].corr(df['Panic_numeric']):.4f}")
```
