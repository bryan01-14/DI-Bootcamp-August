## 🌟 Exercise 2: LLM Architecture and Application Scenarios
For each, describe (a) core architectural differences, (b) a real application, (c) why it fits.

### BERT
- **Architecture**: BERT is an encoder-only transformer that reads the entire input sequence bidirectionally. It is pre-trained using Masked Language Modeling (MLM), where random tokens are masked and the model must predict them using both left and right context. Early versions also used Next Sentence Prediction (NSP). Because it has access to full context in both directions simultaneously, it builds deep contextual representations of each token.
- **Application**: Named Entity Recognition (NER) in medical records — identifying patient names, drug names, diagnoses, and dosages in clinical texts.
- **Why it fits**: NER requires understanding the role of each word within its full sentence context (e.g., distinguishing "Apple" the company from "apple" the fruit). BERT's bidirectional attention allows it to leverage both preceding and following words when labeling each token, making it highly accurate for classification tasks at the token or sentence level.

### GPT
- **Architecture**: GPT is a decoder-only transformer trained with Causal Language Modeling (CLM), also called autoregressive language modeling. It processes tokens from left to right only (unidirectional), using masked self-attention to prevent each token from attending to future tokens. This design makes it naturally suited for sequential text generation.
- **Application**: An AI coding assistant (e.g., GitHub Copilot) that auto-completes code snippets as the developer types.
- **Why it fits**: Code completion is inherently a left-to-right generation task — the model must predict the next token given all previous tokens. GPT's autoregressive nature makes it ideal for generating coherent, contextually relevant continuations of any prompt, whether code or natural language.

### T5
- **Architecture**: T5 (Text-to-Text Transfer Transformer) uses a full encoder-decoder architecture. The encoder processes the input and builds a contextual representation; the decoder then generates the output token by token conditioned on that representation. Every NLP task is framed as converting one text string into another (text-to-text), using task-specific prefixes like "summarize:", "translate English to French:", etc.
- **Application**: Automatic summarization of long scientific papers into concise abstracts.
- **Why it fits**: Summarization requires first deeply understanding the full input (encoder's role) and then generating a coherent, compressed output (decoder's role). The encoder-decoder separation allows T5 to handle long input sequences while producing fluent, accurate summaries — making it a strong fit for any sequence-to-sequence transformation task.


## 🌟 Exercise 3: Benefits and Ethics of Pre-training

### Benefits of Pre-trained Models

1. **Improved generalization**: Pre-training on massive and diverse text corpora forces the model to learn broad linguistic patterns, world knowledge, and semantic relationships that go far beyond any single task. When fine-tuned, the model carries this general knowledge, allowing it to generalize well even to unseen examples in the target domain.

2. **Reduced need for labeled data**: Pre-training is self-supervised — it uses raw, unlabeled text (e.g., Wikipedia, web pages) to learn language representations. As a result, when adapting the model to a downstream task, only a small amount of task-specific labeled data is needed to achieve strong performance, greatly reducing annotation costs.

3. **Faster fine-tuning**: Since the model already has a strong linguistic foundation from pre-training, fine-tuning requires far fewer training steps and epochs compared to training from scratch. This dramatically speeds up the development cycle for new NLP applications.

4. **Transfer learning**: The knowledge and representations learned during pre-training transfer effectively across very different tasks (e.g., a model pre-trained on general web text can be fine-tuned for medical NER, legal classification, or customer sentiment analysis). This "transfer" means one pre-trained model can be the starting point for dozens of applications.

5. **Robustness**: Models pre-trained on diverse data sources tend to be more robust to variations in writing style, vocabulary, typos, and domain shifts. They have seen a wide range of linguistic patterns during pre-training, making them less brittle than models trained on narrow, task-specific datasets.

### Ethical Concerns

- **Bias**: Training data scraped from the internet reflects societal biases (gender, racial, cultural). The model absorbs and can amplify these biases in its outputs — for example, associating certain professions with specific genders.
- **Misinformation**: LLMs can generate convincing but factually incorrect text (hallucinations). If deployed at scale, they can spread false information rapidly.
- **Misuse**: Pre-trained models can be fine-tuned to generate phishing emails, propaganda, fake reviews, or deepfake text, lowering the barrier to malicious content creation.
- **Privacy**: Web-scraped training data may inadvertently include personal information (names, emails, addresses), which the model might memorize and reproduce.

### Mitigation Strategies

- **Data curation & filtering**: Carefully curate training data to remove toxic content, hate speech, and personally identifiable information before pre-training.
- **Differential privacy**: Apply differential privacy techniques during training to limit the model's ability to memorize specific data points.
- **RLHF (Reinforcement Learning from Human Feedback)**: Use human feedback to align model outputs with human values, reducing harmful or biased generations.
- **Safety filters & guardrails**: Deploy content moderation layers at inference time to detect and block harmful outputs.
- **Bias audits**: Regularly evaluate models using fairness benchmarks across demographic groups and publish audit results for transparency and accountability.


## 🌟 Exercise 4: Transformer Architecture Deep Dive

### Self-Attention & Multi-Head Attention

**How self-attention works (Q/K/V flow):**
Each input token embedding is linearly projected into three vectors: a Query (Q), a Key (K), and a Value (V) using learned weight matrices. The attention score between two tokens is computed as the dot product of their Q and K vectors, scaled by the square root of the key dimension (√d_k) to prevent vanishing gradients. These scores are then passed through a softmax function to produce a probability distribution (attention weights) that sums to 1. Finally, these weights are used to compute a weighted sum of all Value vectors, producing the output representation for each token. Tokens with high attention weights contribute more to the final output, effectively allowing each token to "focus" on the most relevant parts of the sequence.

**Why multi-head attention helps:**
Instead of running a single attention function, multi-head attention runs h parallel attention heads, each with its own Q, K, V projection matrices. Each head learns to attend to different aspects or relationships in the input — one head might capture syntactic dependencies, another semantic similarity, another long-range coreference. The outputs of all heads are concatenated and projected back to the original dimension, giving the model a richer, multi-faceted representation of each token than a single head could provide.

**Example sentence and head specialization:**

Sentence: *"The nurse handed the patient her prescription after reviewing the chart."*

- **Head 1 (coreference resolution)**: Might learn that "her" refers back to "nurse" (not "patient"), resolving pronoun-antecedent relations across the sentence.
- **Head 2 (syntactic dependency)**: Might focus on the verb-object relation, linking "handed" to "prescription" and "patient", capturing the action structure.

### Pre-training Objectives

**MLM vs CLM:**
- **MLM (Masked Language Modeling)**, used by BERT: Random tokens in the input are replaced with a [MASK] token, and the model must predict the original token using both left and right context simultaneously. This trains bidirectional representations. It is best suited for understanding tasks (classification, NER, Q&A) where full context is available.
- **CLM (Causal Language Modeling)**, used by GPT: The model predicts the next token given all previous tokens, attending only to past context (left-to-right). This is naturally suited for text generation tasks where output must be produced token by token.

**When to prefer each:**
- Prefer **MLM** for tasks requiring deep understanding of input: sentiment analysis, text classification, extractive Q&A, NER.
- Prefer **CLM** for tasks requiring open-ended generation: chatbots, story generation, code completion, translation (when combined with encoder).

**Next Sentence Prediction (NSP):**
Early BERT used NSP as an auxiliary pre-training task: given two sentence segments, the model predicts whether the second sentence naturally follows the first. The motivation was to train the model to understand inter-sentence relationships for tasks like Natural Language Inference and Q&A. However, subsequent research (RoBERTa, ALBERT) showed that NSP provides little benefit and can actually hurt performance because the task is too easy (negative pairs are often drawn from different documents, making them trivially distinguishable). Modern models dropped NSP and achieved better results by simply training longer with more data using MLM alone.

### Transformer Model Selection

- **Sentiment analysis on customer reviews → Encoder-only (e.g., BERT, RoBERTa)**
  Justification: Sentiment classification is an understanding task — the model reads the full review and outputs a single label. Encoder-only models are ideal because bidirectional attention gives them access to the complete context of each token, producing the richest possible sentence-level representation for classification. No text generation is required.

- **Creative conversational chatbot → Decoder-only (e.g., GPT-3/4)**
  Justification: A chatbot must generate fluent, creative, open-ended responses token by token. Decoder-only models are trained with CLM, making them naturally suited for autoregressive generation. They can produce diverse and contextually coherent responses conditioned on the full conversation history.

- **Technical document translation (EN → ES) → Encoder-Decoder (e.g., T5, mBART)**
  Justification: Translation is a sequence-to-sequence task. The encoder reads and deeply processes the full English source document; the decoder generates the Spanish translation token by token conditioned on the encoder's representation. The explicit encoder-decoder split is designed precisely for tasks where input and output are different sequences of potentially different lengths.

### Positional Encoding

**Purpose:**
The self-attention mechanism is permutation-invariant — it treats the input as a set, not a sequence. Without positional information, the model cannot distinguish between "The dog bit the man" and "The man bit the dog". Positional encoding injects information about the absolute or relative position of each token into its embedding, allowing the model to understand word order, which is fundamental to meaning in most languages.

**Example failure without positional encoding:**
Consider the sentence pair: "Not all students who passed the exam were happy" vs "All students who passed the exam were not happy." Without positional encoding, both sentences would produce identical attention patterns since the same tokens appear in different orders, causing the model to assign the same meaning to two sentences with opposite interpretations. This would lead to completely wrong outputs in tasks like sentiment analysis or natural language inference.


## 🌟 Exercise 5: BERT Variations: Choose Your Detective

### Scenario Assignments

- **Scenario 1 (mobile app, limited resources — real-time sentiment analysis):** **DistilBERT**
  DistilBERT is 40% smaller and 60% faster than BERT while retaining ~97% of its performance. Its compact size makes it ideal for deployment on mobile devices with limited RAM, CPU, and battery.

- **Scenario 2 (legal documents, high accuracy):** **RoBERTa**
  RoBERTa was trained on significantly more data, with larger batches, longer sequences, and without NSP. These improvements lead to stronger and more reliable representations, making it the best choice when maximum accuracy is critical, as in legal document analysis.

- **Scenario 3 (global customer support, multiple languages):** **XLM-RoBERTa (XLM-R)**
  XLM-R is pre-trained on 100 languages using a massive multilingual corpus. It achieves strong cross-lingual transfer, enabling it to handle customer queries in dozens of languages with a single model.

- **Scenario 4 (efficient pre-training, token replacement detection):** **ELECTRA**
  ELECTRA uses a generator-discriminator setup: the generator (a small MLM model) replaces tokens, and the discriminator (the main ELECTRA model) must detect which tokens were replaced. This "replaced token detection" objective is more sample-efficient than MLM — the model learns from every input token, not just the 15% masked ones.

- **Scenario 5 (efficient NLP, resource-constrained environments):** **ALBERT**
  ALBERT uses parameter sharing across layers and factorized embedding parameterization to dramatically reduce its parameter count while maintaining performance. This makes it highly memory-efficient and suitable for resource-constrained environments.

## 🌟 Exercise 6: Softmax Temperature: The Randomness Regulator

### 1) Temperature Scenarios

- **T = 0.2 (very low temperature):** The probability distribution becomes extremely peaked — the model almost always picks the single highest-probability token at each step. Output is highly deterministic, repetitive, and conservative. The text tends to be safe and predictable but can become monotonous and lack creativity. Useful when precision and reliability are paramount (e.g., factual Q&A).

- **T = 1.5 (high temperature):** The distribution becomes very flat — many tokens get similar probabilities, including low-probability (surprising) ones. Output becomes highly diverse, creative, and sometimes incoherent. The model may produce unusual word choices, unexpected sentence structures, or logically inconsistent text. Useful for brainstorming or creative writing where novelty is valued over accuracy.

- **T = 1.0 (neutral temperature):** The model uses the raw softmax probabilities as learned during training, without sharpening or flattening. This is the baseline behavior: reasonably coherent and moderately diverse. A good default starting point before tuning for specific use cases.

### 2) Application Design

- **Bedtime stories (creativity vs coherence):** Use a **moderate temperature between 0.7 and 1.0**, potentially with top-p sampling (nucleus sampling, p ≈ 0.9). This keeps the story coherent and grammatically correct (avoiding the chaos of T > 1.2) while still allowing creative vocabulary, unexpected plot twists, and vivid imagery. A temperature that is too low would produce formulaic, boring stories; too high would generate incoherent or frightening content for children.

- **Financial report summaries (accuracy/reliability):** Use a **very low temperature (T ≈ 0.1 to 0.3)**, ideally combined with greedy decoding or beam search. Financial summaries require factual accuracy — the model must reproduce numbers, trends, and conclusions faithfully from the source document. High temperature risks introducing hallucinated figures or misrepresented facts, which could be legally and financially consequential.

### 3) Temperature & Bias

Temperature can both surface and suppress biases embedded in a model's training data:

- **Low temperature amplifies dominant biases**: At T = 0.2, the model almost always picks its most probable prediction — which reflects the most frequent patterns in training data. If the training data associates "nurse" predominantly with female pronouns, the model at low temperature will almost always generate "she" when referring to a nurse, consistently surfacing this gender bias.

- **High temperature partially masks biases**: At T = 1.5, the distribution is flatter, so lower-probability (less biased) tokens get a chance to be selected. This may reduce the consistency of biased outputs, though it does not eliminate the underlying bias — it simply adds noise.

**Concrete example:** A recruitment assistant LLM is asked to complete: *"The ideal software engineer is someone who..."*
- At T = 0.2: The model consistently generates stereotypically male-coded traits ("he is logical, competitive, works long hours"), faithfully reproducing the dominant pattern in tech job description data.
- At T = 1.0: More varied completions appear, some neutral, some still biased.
- At T = 1.5: The outputs vary widely — some neutral, some creative — but the bias is not reliably eliminated, just randomized.

This illustrates that temperature tuning alone is insufficient to address bias; proper data curation, debiasing techniques, and RLHF are necessary.


### (Optional) Quick Generation Demo Across Temperatures
Note: Requires `transformers` and model download; skip if offline.


```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

model_name = "gpt2"  # Causal LM for text generation
prompt = "In the future, cybersecurity will"  # Custom prompt related to our field

pipe = pipeline(
    "text-generation",
    model=AutoModelForCausalLM.from_pretrained(model_name),
    tokenizer=AutoTokenizer.from_pretrained(model_name),
    device=0 if torch.cuda.is_available() else -1,
)

for T in [0.2, 1.0, 1.5]:
    out = pipe(prompt, max_new_tokens=40, temperature=T, do_sample=True)
    print("--- Temperature:", T)
    print(out[0]["generated_text"])  # Observe how style and creativity shift with T
    print()
```
